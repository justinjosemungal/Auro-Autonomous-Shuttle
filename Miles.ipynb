```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
miles = pd.read_csv('Miles.csv', header = None, sep=',')
miles.head()
```

```python
miles.columns = ['id', 'mode', 'start_timestamp', 'end_timestamp', 'miles']
```

```python
miles.start_timestamp = pd.to_datetime(miles.start_timestamp)
miles.end_timestamp = pd.to_datetime(miles.end_timestamp)
```

```python
temp = pd.DatetimeIndex(miles.start_timestamp)
miles['start_date'] = temp.date
miles['start_time'] = temp.time
temp = pd.DatetimeIndex(miles.end_timestamp)
miles['end_date'] = temp.date
miles['end_time'] = temp.time
```

```python
miles.to_csv('final_Miles.csv', sep=',', index = False)
```
